## 📌 Tekrar Amaçlı Çalışma / Proje Listesi

### 🔹 Part 1 – ANN (Artificial Neural Networks)

* **Proje:** Banka müşteri kaybı (churn) tahmini
* **Veri:** Kaggle “Bank Customer Churn” dataset
* **Amaç:** Eğitim setini temizleyip, ANN ile “müşteri kalır mı/ayrılır mı” sınıflandır.

---

### 🔹 Part 2 – CNN (Convolutional Neural Networks)

* **Proje:** Görüntü sınıflandırma
* **Veri:** Kaggle Cats vs Dogs veya CIFAR-10
* **Amaç:** CNN ile hayvan/nesne sınıflandırma, ardından Transfer Learning (VGG/ResNet) dene.

---

### 🔹 Part 3 – RNN / LSTM / GRU

* **Proje:** Zaman serisi tahmini
* **Veri:** Google/Apple hisse fiyatları, veya Türkiye elektrik tüketimi (TÜİK verileri bile olur)
* **Amaç:** LSTM ile gelecek birkaç günün değerini tahmin et, GRU ile karşılaştır.

---

### 🔹 Part 4 – SOM (Self Organizing Maps)

* **Proje:** Müşteri segmentasyonu + anomali tespiti
* **Veri:** Kredi kartı müşteri verileri
* **Amaç:** SOM ile müşteri kümelerini görselleştir, “outlier” olanları fraud adayları olarak işaretle.

---

### 🔹 Part 5 – RBM / Boltzmann Machines

* **Proje:** Film tavsiye sistemi
* **Veri:** MovieLens 100k veya 1M dataset
* **Amaç:** RBM ile kullanıcı-film eşleşmesi, klasik matrix factorization ile kıyasla.

---

### 🔹 Part 6 – AutoEncoders

* **Proje:** Anomali tespiti + recommender
* **Veri:** Kredi kartı fraud dataset / MovieLens
* **Amaç:**

  * (1) Autoencoder ile “rekonstrüksiyon hatası” üzerinden fraud tespiti
  * (2) Autoencoder tabanlı film öneri sistemi

---

## 📌 Bundan Sonra Geçmen Gereken Aşamalar

Şu an **Deep Learning temelleri**ni öğrendin. Bundan sonrası 3 koldan ilerleyebilir:

### 1. **İleri Derin Öğrenme**

* Transformer’lar (BERT, GPT, Vision Transformer)

* Flask/FastAPI + Docker ile model servis etme

* GPU optimizasyonu, model deployment (AWS, GCP, on-premises)
