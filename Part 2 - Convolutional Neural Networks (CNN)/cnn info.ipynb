## 1. Neden CNN?

* **Sorun (ANN ile görüntü tanıma):**
  Diyelim elimizde **100x100 piksellik** siyah-beyaz bir resim var.

  * Girdi sayısı = 10,000 (100\*100).
  * Eğer bu resmi doğrudan ANN’e verirsek, her piksel bir input node olur → devasa sayıda ağırlık parametresi çıkar.
  * Bu hem **çok yavaş**, hem de **overfitting riski** yüksek.

* **Çözüm (CNN ile):**
  Görüntüdeki **uzaysal yapıyı (spatial structure)** koruyan özel katmanlar (convolution + pooling) ekleyelim.
  Böylece model:

  * Kenarları, köşeleri, desenleri, dokuları **otomatik öğrenir.**
  * Daha az parametreyle çok daha güçlü olur.

## 2. CNN Mimarisi (Büyük Resim)

Bir CNN genelde şu katmanlardan oluşur:

1. **Convolution Layer (Evrişim Katmanı):**

   * Görüntü üzerine küçük filtreler (kernel) kaydırılır.
   * Amaç: Kenar, çizgi, köşe gibi temel özellikleri yakalamak.
   * Örn: İlk katmanda kenarları öğrenir, sonraki katmanda şekilleri, daha sonraki katmanda objeleri.

2. **ReLU Activation:**

   * Negatif değerleri sıfırlar, pozitifleri bırakır.
   * Modele doğrusal olmayanlık (non-linearity) katar.

3. **Pooling Layer (Özellik Azaltma):**

   * En yaygın: **Max Pooling** → küçük bir bölgedeki maksimum değeri alır.
   * Amaç: Boyutu küçültmek (hesaplama hızlanır), gürültüyü azaltmak, **spatial invariance** (bir nesne biraz kaymış olsa da tanınsın).

4. **Flattening:**

   * 2D özellik haritaları tek boyutlu vektöre dönüştürülür.
   * ANN’deki fully connected (dense) katmanlara bağlanır.

5. **Fully Connected (Dense) Layers:**

   * Son kısım klasik ANN gibidir.
   * Özelliklerden tahmin yapılır (ör. “kedi mi köpek mi?”).

---

## 3. CNN’in Gücü

* **Parametre paylaşımı:** Aynı filtre tüm resim üzerinde kaydırılır → çok daha az parametre.
* **Yerel bağlantılar:** Her nöron sadece küçük bir bölgeden bilgi alır → beyne daha yakın bir yapı.
* **Çeviri kaymazlığı (translation invariance):** Nesne resimde sağa kaydırılsa da CNN onu tanıyabilir.

## 4. Sezgisel Örnek

Bir fotoğrafa baktığında:

* İlk bakışta **çizgileri, kenarları** fark edersin (Convolution).
* Daha sonra **daha büyük şekilleri** görürsün (Pooling + üst katmanlar).
* En sonunda **nesneyi tanırsın** (Fully Connected Layer).

CNN de tıpkı böyle çalışır: küçük detaylardan başlayarak soyut seviyeye çıkar.

---

## 5. Özet

* ANN → düz veri (tablo, sayılar).
* CNN → görüntü/veri içindeki **uzaysal ilişkileri** koruyarak işler.
* Yapı taşları: **Convolution → ReLU → Pooling → Flatten → Fully Connected.**
