## 1. Neden CNN?

* **Sorun (ANN ile grnt tanma):**
  Diyelim elimizde **100x100 piksellik** siyah-beyaz bir resim var.

  * Girdi says = 10,000 (100\*100).
  * Eer bu resmi dorudan ANNe verirsek, her piksel bir input node olur  devasa sayda arlk parametresi kar.
  * Bu hem **ok yava**, hem de **overfitting riski** yksek.

* **zm (CNN ile):**
  Grntdeki **uzaysal yapy (spatial structure)** koruyan zel katmanlar (convolution + pooling) ekleyelim.
  Bylece model:

  * Kenarlar, keleri, desenleri, dokular **otomatik renir.**
  * Daha az parametreyle ok daha gl olur.

## 2. CNN Mimarisi (Byk Resim)

Bir CNN genelde u katmanlardan oluur:

1. **Convolution Layer (Evriim Katman):**

   * Grnt zerine kk filtreler (kernel) kaydrlr.
   * Ama: Kenar, izgi, ke gibi temel zellikleri yakalamak.
   * rn: lk katmanda kenarlar renir, sonraki katmanda ekilleri, daha sonraki katmanda objeleri.

2. **ReLU Activation:**

   * Negatif deerleri sfrlar, pozitifleri brakr.
   * Modele dorusal olmayanlk (non-linearity) katar.

3. **Pooling Layer (zellik Azaltma):**

   * En yaygn: **Max Pooling**  kk bir blgedeki maksimum deeri alr.
   * Ama: Boyutu kltmek (hesaplama hzlanr), grlty azaltmak, **spatial invariance** (bir nesne biraz kaym olsa da tannsn).

4. **Flattening:**

   * 2D zellik haritalar tek boyutlu vektre dntrlr.
   * ANNdeki fully connected (dense) katmanlara balanr.

5. **Fully Connected (Dense) Layers:**

   * Son ksm klasik ANN gibidir.
   * zelliklerden tahmin yaplr (r. kedi mi kpek mi?).

## 3. CNNin Gc

* **Parametre paylam:** Ayn filtre tm resim zerinde kaydrlr  ok daha az parametre.
* **Yerel balantlar:** Her nron sadece kk bir blgeden bilgi alr  beyne daha yakn bir yap.
* **eviri kaymazl (translation invariance):** Nesne resimde saa kaydrlsa da CNN onu tanyabilir.
