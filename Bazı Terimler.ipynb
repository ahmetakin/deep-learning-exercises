### Eğitim Parametreleri & Kontroller

* **plateau**: Modelin val_loss’u bir süre iyileşmezse öğrenme oranını düşürmek için kullanılan durum (ReduceLROnPlateau).
* **early (early stopping)**: Val_loss veya seçilen metriği izleyip ilerleme durunca eğitimi erken bitirir.
* **ckpt (checkpoint)**: Eğitimin belirli aşamalarında modelin kaydedilen sürümleri. En iyi sonuç veren model saklanır.
* **history**: Model.fit sonrası, epoch bazında loss ve metrik geçmişini tutan yapı.
* **callbacks**: Eğitim sırasında otomatik müdahaleler (ör: early stopping, model kaydetme, LR düşürme).
* **epochs**: Tüm veri setinin ağa bir kez gösterilmesi. Eğitim turu sayısı.
* **batch_size**: Kaç örneğin aynı anda işlenip ağı güncellediği. Bellek ve hızla ilgili.

---

### Kayıp & Hata Ölçümleri

* **val_loss**: Doğrulama (validation) setinde hesaplanan loss değeri. Overfitting takibi için kritik.
* **mae (Mean Absolute Error)**: Tahmin ile gerçek arasındaki ortalama mutlak fark.
* **ssim (Structural Similarity Index)**: Görüntülerin yapısal benzerliğini ölçer (0–1 arası).
* **PSNR (Peak Signal-to-Noise Ratio)**: Görüntü kalitesi ölçütü, yüksek değer daha iyi.

---

### Aktivasyon & Optimizasyon

* **optimizer**: Ağırlıkları güncelleme algoritması (SGD, Adam, RMSProp vb.).
* **activation**: Katman çıkışını dönüştüren fonksiyon (ReLU, Sigmoid, Tanh, Softmax).

---

### Sınıflandırma Ölçütleri

* **roc_auc_score**: ROC eğrisi altındaki alan. Pozitif/negatif ayrım gücünü ölçer.
* **average_precision_score**: Precision-Recall eğrisinin altında kalan alan. Özellikle dengesiz veri setlerinde önemli.
* **precision_recall_curve**: Eşik değiştikçe precision/recall’un nasıl değiştiğini gösterir.
* **roc_curve**: Eşik değiştikçe TPR/FPR’nin nasıl değiştiğini gösterir.
* **classification_report**: Precision, Recall, F1-score, Support özet tablosu.
* **confusion_matrix**: Tahmin vs Gerçek dağılım tablosu (TP, TN, FP, FN).
* **threshold**: Sınıflandırmada “pozitif” karar eşiği (genelde 0.5).

---

### Ekleyebileceğim Faydalı Terimler

* **loss**: Modelin hata fonksiyonu (MSE, Cross-Entropy vb.).
* **val_accuracy**: Doğrulama setinde elde edilen başarı oranı.
* **dropout**: Overfitting önlemek için eğitimde bazı nöronları rastgele kapatma.
* **learning_rate (lr)**: Ağırlık güncellemelerinin adım büyüklüğü.
* **f1_score**: Precision & Recall’un dengeli ortalaması.
* **TPR / Recall / Sensitivity**: Pozitifleri yakalama oranı.
* **TNR / Specificity**: Negatifleri doğru bulma oranı.
* **precision**: Pozitif tahminlerin ne kadarının doğru olduğu.
* **overfitting**: Modelin eğitim setine fazla uyum sağlayıp genellemede başarısız olması.
* **underfitting**: Modelin veriyi öğrenememesi, hem train hem val kötü.
